# Factor

Los factores son una estructura de datos para manejar variables categóricas en R. Son útiles para representar datos cualitativos y permiten realizar análisis estadísticos de manera más eficiente.

**Variable categorica:** Una variable categórica es aquella que toma un número finito de valores, conocidos como categorías o niveles.
Ejemplo: edad, género, color, dias de la semana, etc.

```r
dias <- c("lunes", "martes", "miercoles", "jueves", "viernes",
          "lunes", "sabado", "domingo", "domingo")
dias_f <- factor(dias)
plot(dias_f)
```

```r
levels(dias_f)
```

Modificar la función `factor()`. LIMPIEZA DE DATOS.

```r
dias_mod <- factor(dias,
                   levels = c("lunes", "martes", "miercoles", "jueves", "viernes", "sabado", "domingo"),
                   labels = c("L", "M", "X", "J", "V", "S", "D"))
plot(dias_mod)
```

Modificar y dar orden los niveles del factor

```r
dias_ord <- factor(dias,
                   ordered = TRUE,
                   levels = c("lunes", "martes", "miercoles", "jueves", "viernes", "sabado", "domingo"),
                   labels = c("L", "M", "X", "J", "V", "S", "D"))
dias_ord
plot(dias_ord)
```
